```python
import pyabf
import matplotlib.pyplot as plt

def plot_first_sweep(abf_path):
    """
    Plots the first sweep (sweep 0) of an ABF file.
    """
    abf = pyabf.ABF(abf_path)
    abf.setSweep(0)

    time_ms = abf.sweepX * 1000  # Convert from seconds to milliseconds
    current_pA = abf.sweepY

    plt.figure(figsize=(10, 5))
    plt.plot(time_ms, current_pA, color='black', lw=1)
    plt.title(f"{abf.abfID} – Sweep 0")
    plt.xlabel("Time (ms)")
    plt.ylabel(f"Current ({abf.sweepUnitsY})")
    plt.xlim([800, 1200])
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
