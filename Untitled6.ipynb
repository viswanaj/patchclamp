```python
import os
import pyabf
import numpy as np
import matplotlib.pyplot as plt

def analyze_rmp(abf_file, time_window=[200, 220]):
    """
    Analyze the resting membrane potential (RMP) from an ABF file.

    Parameters:
        abf_file (str): Path to the ABF file.
        time_window (list): Time window [start, end] in ms for RMP calculation.

    Returns:
        float: Average RMP across sweeps or None if the file is not valid.
    """
    abf = pyabf.ABF(abf_file)

    if "mV" not in abf.sweepLabelY:
        print(f"Skipping {abf_file}: Not a current clamp recording.")
        return None  

    resting_membrane_potentials = []

    for sweep in range(abf.sweepCount):
        abf.setSweep(sweep)
        time_idx = (abf.sweepX * 1000 >= time_window[0]) & (abf.sweepX * 1000 <= time_window[1])
        rmp = np.mean(abf.sweepY[time_idx])
        resting_membrane_potentials.append(rmp)

    avg_rmp = np.mean(resting_membrane_potentials)
    print(f"Processed {abf_file}: Avg RMP = {avg_rmp:.2f} mV")

    return avg_rmp

def process_current_clamp_files(directory, time_window=[200, 220], plot=True):
    """
    Identify and analyze all current clamp ABF files in a directory.

    Parameters:
        directory (str): The path to the folder containing ABF files.
        time_window (list): Time window [start, end] in ms for RMP calculation.
        plot (bool): Whether to generate a plot.

    Returns:
        dict: {filename: (protocol_name, avg_rmp)}
    """
    results = {}

    for file in os.listdir(directory):
        if file.endswith(".abf"):
            file_path = os.path.join(directory, file)
            abf = pyabf.ABF(file_path)

            # Check if it's a current clamp file
            if "mV" in abf.sweepLabelY:
                protocol_name = abf.protocol if abf.protocol else "Unknown"
                avg_rmp = analyze_rmp(file_path, time_window)

                if avg_rmp is not None:
                    results[file] = (protocol_name, avg_rmp)

    # Save results to a text file
    output_path = os.path.join(directory, "rmp_results.txt")
    with open(output_path, "w") as f:
        for filename, (protocol, avg_rmp) in results.items():
            f.write(f"{filename} ({protocol}): {avg_rmp:.2f} mV\n")

    print(f"Analysis complete. Results saved to {output_path}")

    # Plot results if needed
    if plot and results:
        plot_rmp_trends(results, directory)

    return results

def plot_rmp_trends(results, directory):
    """
    Plot the resting membrane potential (RMP) across multiple ABF files.

    Parameters:
        results (dict): {filename: (protocol, avg_rmp)}
        directory (str): Directory to save the plot.
    """
    filenames = list(results.keys())
    rmp_values = [rmp for _, rmp in results.values()]
    protocols = [protocol for protocol, _ in results.values()]

    plt.figure(figsize=(8, 5))
    plt.scatter(filenames, rmp_values, color='b', label="Avg RMP")
    plt.axhline(np.mean(rmp_values), color='r', linestyle='--', label="Mean RMP")

    plt.xlabel("ABF Files")
    plt.ylabel("Resting Membrane Potential (mV)")
    plt.title("Resting Membrane Potential Across Files")
    plt.xticks(rotation=45, ha='right')
    plt.legend()
    plt.grid(True)

    # Save and show the plot
    plot_path = os.path.join(directory, "rmp_plot.png")
    plt.savefig(plot_path, bbox_inches="tight")
    plt.show()

    print(f"Plot saved as {plot_path}")
```

```python
data_directory = "/Users/jayashri/Desktop/03232025"
time_window = [200, 220]  # Adjustable time window

# Process all current clamp files and generate a plot
results = process_current_clamp_files(data_directory, time_window, plot=True)

# Print results
for file, (protocol, avg_rmp) in results.items():
    print(f"{file} ({protocol}): {avg_rmp:.2f} mV")
```
